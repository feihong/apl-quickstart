Dfns that take multiple lines must be defined using ]dinput. Note that cells using ]dinput cannot have comments before the opening curly brace, and they can't have anything else in them other than the dfn definition.

```apl
]dinput
RandInt ← {
  ⍝ Left argument gets a default value if it's not provided
  ⍺ ← 1
  ⍺+?(1+⍵-⍺)
}
```

```apl
RandInt 4
```

```apl
⎕RL ← 73
RandInt 5⊣¨⍳32
5 RandInt 9⊣¨⍳32
```

```apl
⎕RL ← 888
RandomHanzi 32
```

You can also define tradfns, which can use traditional programming constructs. Note that they have dynamic scoping, unlike dfns, which use lexical scoping. You must declare all local variables you use in the header, otherwise variables in the tradfn body will become global variables.

```apl
∇r ← {start} RandIntTrad end; count
  :If 900⌶⍬
    start ← 1 ⍝ default value if this function is called monadically
  :EndIf
  count ← (1+end-start) ⍝ this would be global if it wasn't previously declared
  r ← start+?count
∇
```

```apl
⎕RL ← 73
RandInt 5⊣¨⍳32
5 RandIntTrad 9⊣¨⍳32
```
