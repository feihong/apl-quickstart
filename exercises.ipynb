```apl
]dinput
Frame ← {
  (r c) ← ⍴⍵
  hline ← c⍴⎕UCS 9472
  inner ← hline⍪⍵⍪hline
  vline ← r⍴⎕UCS 9474
  left ← (⎕UCS 9484),vline,⎕UCS 9492
  right ← (⎕UCS 9488),vline,⎕UCS 9496
  left,inner,right
}
⍝ 5.7
```

```apl
]dinput
Frame ← {
  shape ← ⍴⍵
  (r c) ← ((1=≢shape)⍴1),shape
  mat ← r c⍴⍵
  hline ← c⍴⎕UCS 9472
  inner ← hline⍪mat⍪hline
  vline ← r⍴⎕UCS 9474
  left ← (⎕UCS 9484),vline,⎕UCS 9492
  right ← (⎕UCS 9488),vline,⎕UCS 9496
  left,inner,right
}
⍝ 5.8
```

```apl
]dinput
Switch ← {
  (x y) ← ⍺
  s ← ⍵,y
  pos ← ⍳≢s
  replaceXWith0 ← pos×x≠s
  replaceXWithY ← (x=s)×s⍳y
  ¯1↓s[replaceXWith0+replaceXWithY]
}
⍝ 5.9
```

```apl
'tc' Switch 'A bird in the hand is worth two in the bush'
```

```apl
]dinput
Swap ← {
  (x y) ← ⍺
  s ← ⍵,x,y
  pos ← ⍳≢s
  replaceXYWith0 ← pos×~s∊x y
  replaceXWithY ← (x=s)×s⍳y
  replaceYWithX ← (y=s)×s⍳x
  ¯2↓s[replaceXYWith0+replaceXWithY+replaceYWithX]
}
⍝ 5.10
```

```apl
'ei' Swap 'A bird in the hand is worth two in the bush'
```
