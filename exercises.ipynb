```apl
]dinput
Reverse ← {
  count ← ≢⍵
  ⍵[1+count-⍳count]
}
⍝ 5.3
```

```apl
]dinput
Totalise ← {
  withRowSums ← ⍵,+/⍵
  withRowSums,[1]+⌿withRowSums
}
⍝ 5.4
```

```apl
mat ← 3 4⍴75 14 86 20 31 16 40 51 22 64 31 28
Totalise mat
```

```apl
]dinput
Lengths ← {
  pos ← ⍸' '=' ',⍵,' '
  ¯1+(1↓pos)-¯1↓pos
}
⍝ 5.5
```

```apl
Lengths 'This seems to be a good solution'
Lengths 'hello'
```

```apl
⍝ 5.6
To ← {(⍺-1)+⍳1+⍵-⍺}
17 To 29
55 To 55
```

```apl
]dinput
Frame ← {
  (r c) ← ⍴⍵
  hline ← c⍴'-'
  inner ← hline⍪⍵⍪hline
  vline ← '+',(r⍴'|'),'+'
  vline,inner,vline
}
⍝ 5.7
```
